```python
def generate_machines(num_nodes, seed):
    '''
    Generates a random list of num_nodes machines. Each machine is a list of
    its attributes (cpu speed, memory size, power budget). Seed is for
    reproducing machines if needed.
    '''
    random.seed(seed)
    machine_list = []
    for i in range(num_nodes):
        cpu = random.randrange(20, 50) / 10 #Change to remove fractions
        mem = random.choice([8, 16, 32, 64])
        power = random.randrange(20, 50) / 10
        machine_list.append([cpu, mem, power])
    return machine_list
```

```python
def comp_score(task_req, machine):
    '''
        Calculates computation performance score, score represents the task's
        performance on its assigned machine (lower is better).
    '''
    score = task_req / (machine[0] * 10**9)
    return score


def commu_score(task_reqs, mapping, machine_graph, machine_sender):
    """
        Calculates communication performance score, score represents the task's
        performance on its assigned machine (lower is better).
        Communication score currently assumes all links are 10 GB/s
    """
    distances = dijkstra(machine_graph, machine_sender)
    score = 0
    for i in task_reqs:
        #print(i)
        #print("sending machine: " + str(machine_sender))
        try:
            mapping.index(i[0])
        except:
            score = math.inf
            return score
        score += distances[mapping.index(i[0])] * i[1] / (10 * 10**9)
    return score


def evaluate_map(mapping, machines, machine_graph, calculation, communication):
    '''
        This function will calculate a score for a mapping using
        the computation score and communication score. A score
        of 0 or less indicates an invalid mapping, due to either
        violating a constraint such as power draw, or due to
        some other issue, such as two copies of the same application
        being mapped. The return is the inverse of the summed
        computation and communication scores.
        TODO: Fix power constraint
        TODO: Fix latency constraint
    '''
    total_score = 0
    for item in mapping:    # checking for any duplicates of tasks in mapping
        if mapping.count(item) > 1:
            return -math.inf
    for j, i in enumerate(mapping):
        calc_score = 0
        comm_score = 0
        calc_score = comp_score(calculation[i][i], machines[i])
        receivers = [[x, y] for x, y in enumerate(communication[i]) if y!= 0]
        comm_score = commu_score(receivers, mapping, machine_graph, j)
        score_machine = calc_score + comm_score
        total_score += score_machine
    if total_score == 0:
      return 0
    else:
      return 1/total_score


def dijkstra(graph, start):
    '''
        Implementation of Dijkstra's algorithm taken from
        https://www.algorithms-and-technologies.com/dijkstra/python
    '''
    distances = [float("inf") for _ in range(len(graph))]
    visited = [False for _ in range(len(graph))]
    distances[start] = 0
    while True:
        shortest_distance = float("inf")
        shortest_index = -1
        for i in range(len(graph)):
            if distances[i] < shortest_distance and not visited[i]:
                shortest_distance = distances[i]
                shortest_index = i
        if shortest_index == -1:
            return distances
        for i in range(len(graph[shortest_index])):
            if graph[shortest_index][i] != 0 and distances[i] > distances[shortest_index] + graph[shortest_index][i]:
                distances[i] = distances[shortest_index] + graph[shortest_index][i]
        visited[shortest_index] = True


def mutate(old_map):
    '''
        Mutation function that will swap the assigned machine of two randomly
        chosen tasks in the mapping without affected the original mapping.
    '''
    swap_indices = range(len(old_map))
    new_map = copy.deepcopy(old_map)
    first_index, second_index = random.sample(swap_indices, 2)
    new_map[first_index], new_map[second_index] = new_map[second_index], new_map[first_index]
    return new_map

def middle_factors(n):
  '''
      When given an integer, will return the middle factor pair.
      Ex: 32 returns [4,8], 64 returns [8,8]
  '''
  factor_list = []
  i = 1
  while i*i <= n:
    if n % i == 0:
      factor_list.append([i,n//i])
    i += 1
  return factor_list[-1]

def make_mesh(n):
  middle = middle_factors(n)
  rows = middle[0]
  cols = middle[1]
  n = rows*cols
  M = [[0 for _ in range(n)] for _ in range(n)]
  for r in range(rows):
    for c in range(cols):
      i = r*cols + c
      # Two inner diagonals
      if c > 0: 
        M[i-1][i] = 1
        M[i][i-1] = 1
      # Two outer diagonals
      if r > 0: 
        M[i-cols][i] = 1
        M[i][i-cols] = 1
  return M


def generate_graph(num_nodes, seed):
    '''
    Generates a random graph of machines in a network, represented by an adj
    matrix. Seed is for reproducing machines if needed.
    '''
    random.seed(seed)
    graph = [[0 for i in range(num_nodes)] for j in range(num_nodes)]
    for i in range(num_nodes):
        for j in range(num_nodes):
            if i != j:
                if random.randrange(0, 3) == 1: #33% chance of connection
                    graph[i][j] = 1
                    graph[j][i] = 1
    for i in range(num_nodes):
        if 1 not in graph[i]:
            element = random.randrange(0, num_nodes)
            graph[i][element] = 1
            graph[element][i] = 1
    return graph
```

```python
def sim_anneal(machines, machine_graph, calculation, communication):
    '''
        Simulated annealing function built from descriptions in
        Jong-Kook Kim's thesis. Currently all initial variables
        come from the results in the paper. This function will
        use the simulated annealing technique to find a well 
        performing mapping.
    '''
    temp = 10000
    epsilon = 0.5
    alpha = 0.95
    tasks = [i for i in range(len(calculation))]
    if len(tasks) >= len(machine_graph):
      old_map = random.sample(tasks, len(machine_graph))
    else:
      old_map = random.sample(tasks, len(tasks))
    #print(old_map)
    iters = 0
    while(iters < 250 or temp < 1e-20):
        new_map = mutate(old_map)
        new_eval = evaluate_map(new_map, machines, machine_graph, calculation, communication)
        old_eval = evaluate_map(old_map, machines, machine_graph, calculation, communication)
        #print(old_eval)
        #print(new_eval)
        epsilon_check = epsilon < math.e ** (-abs(new_eval - old_eval) / temp)
        if (new_eval > old_eval) or epsilon_check:
            old_map = new_map
        temp = temp * alpha
        iters += 1
    return new_map


#anneal_map = sim_anneal(machines_list, machines_graph, tasks_calc[:7], tasks_comm[:7])
#evaluation = evaluate_map(anneal_map, machines_list, machines_graph, tasks_calc[:7], tasks_comm[:7])
#print(f"Simulated annealing found: {anneal_map}. Score: {evaluation}.")
```

```python
def makeWheel(population, probability):
    '''
        makeWheel, binSearch, and select are used in the genetic
        algorithm to simulate a roulette wheel. Each slot is a
        mapping and each mapping can have more slots if it scores
        well in the evaluation function. This results in a sampler
        that takes the fitness into account.
    '''
    wheel = []
    total = sum(probability)
    top = 0
    for i in range(len(population)):
        f = probability[i] / total
        wheel.append((top, top + f, population[i][0]))
        top += f
    return wheel


def binSearch(wheel, num):
    mid = len(wheel)//2
    low, high, answer = wheel[mid]
    if low <= num <= high:
        return answer
    elif high < num:
        return binSearch(wheel[mid+1:], num)
    else:
        return binSearch(wheel[:mid], num)


def select(wheel, N):
    stepSize = 1.0/N
    answer = []
    r = random.random()
    answer.append(binSearch(wheel, r))
    while len(answer) < N:
        r += stepSize
        if r > 1:
            r %= 1
        answer.append(binSearch(wheel, r))
    return answer


def eval_genetic(population, machines, machine_graph, calculation, communication, pop_size, n_max, n_min):
    '''
        This is used in the genetic algorithm to sort the mappings
        based on their evaluation score. The probability calculation
        comes from equation 2.4 on p.g. 27 of the thesis.
    '''
    sorted_mappings = []
    for i in range(len(population)):
        map_eval = evaluate_map(population[i], machines, machine_graph, calculation, communication)
        sorted_mappings.append([population[i], map_eval])
    sorted_mappings.sort(key=lambda a: a[1], reverse=True)
    map_probability = []
    for i in range(len(population)):
        prob = (1 / pop_size) * (n_max - (n_max - n_min) * (i / (pop_size - 1)))
        map_probability.append(prob)
    return (sorted_mappings, map_probability)


def genetic(machines, machine_graph, calculation, communication):
    '''
        Genetic algorithm function built from descriptions in
        Jong-Kook Kim's thesis. Currently all initial variables
        come from the results in the paper. This function will
        use the genetic algorithm technique to find a well-
        performing mapping.
    '''
    '''
        # TODO: seed initial pop with duplex (choose best from MCTF and MCPF)
    '''
    cross_over = 0.9
    swap_prob = 0.5
    mutation_rate = 0.05
    stop_generation = 200
    pop_size = 100
    n_max = 1.1
    n_min = 0.9
    assert n_max + n_min == 2, "n_max + n_min must equal 2."
    assert n_max >= 1 and n_max <= 2, "n_max must be inclusively between 1 and 2."
    tasks = [i for i in range(len(calculation))]
    num_machines = len(machine_graph)
    if len(tasks) >= num_machines:
      perm = list(permutations(tasks, num_machines))
    else:
      perm = list(permutations(tasks, len(tasks)))
    population = random.sample(perm, pop_size)
    for i in range(len(population)):
        population[i] = list(population[i])
    
    # seed population here
    
    eval_tuple = eval_genetic(population, machines, machine_graph, calculation, communication, pop_size, n_max, n_min)
    sorted_mappings = eval_tuple[0]
    map_probability = eval_tuple[1]
    generation = 0
    while(generation < stop_generation):
        best_chromosome = sorted_mappings[0]
        # selection
        mating_pool = select(makeWheel(sorted_mappings, map_probability), pop_size)
        # crossover
        children = []
        for i in range(pop_size):
            mating_pair = random.sample(mating_pool, 2)
            r = random.random()
            if r <= 0: # cross_over
                for j in range(len(mating_pair[0])):
                    if random.random() < swap_prob:
                        mating_pair[0][j], mating_pair[1][j] = mating_pair[1][j], mating_pair[0][j]
            children.append(mating_pair[0])
            children.append(mating_pair[1])
        # mutation
        for x in range(len(children)):
            r = random.random()
            if r <= mutation_rate:
                mutate(children[x])
        # evaluate
        eval_tuple = eval_genetic(children, machines, machine_graph, calculation, communication, pop_size, n_max, n_min)
        sorted_mappings = eval_tuple[0]
        map_probability = eval_tuple[1]
        # elitism
        if(sorted_mappings[0][1] < best_chromosome[1]):
            sorted_mappings[0] = best_chromosome
        generation += 1
    return sorted_mappings[0]


#genetic(machines_list, machines_graph, tasks_calc[:7], tasks_comm[:7])
```
